```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
data= pd.read_csv('Milan.csv') 
data
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Rellenar datos nulos con el valor de media de la columna, y con función forward fill en los datos de texto (fechas)
data["last_review"]=data["last_review"].fillna(method="bfill")
data["reviews_per_month"]=data["reviews_per_month"].fillna(round(data["reviews_per_month"].mean(),3))
data
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Eliminamos las clumnas de license y neighborhood group debido a la alta cantidad de outliers, por lo cual consideramos las columnas no muy útiles. 
data=data.dropna(1)
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data.to_csv("MilanLimpio")
```
