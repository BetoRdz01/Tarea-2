```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
data= pd.read_csv('Montreal.csv') 
data
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Rellenar datos nulos con el valor de media de la columna, y con función forward fill en los datos de texto (fechas)
data["last_review"]=data["last_review"].fillna(method="ffill")
data["reviews_per_month"]=data["reviews_per_month"].fillna(round(data["reviews_per_month"].mean(),3))
data["name"]=data["name"].fillna("Desconocido")
data["host_name"]=data["host_name"].fillna("Desconocido")
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data2= data.dropna(1)
data2
```

```python
#Identificar valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
data2.to_csv("MontrealLimpio")
```
